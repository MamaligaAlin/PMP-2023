```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def posterior_grid(grid_points=50, heads=60, tails=90):
    grid = np.linspace(0, 1, grid_points)
    prior = (grid <= 0.5).astype(int)

    likelihood = stats.binom.pmf(heads, heads + tails, grid)
    posterior = likelihood * prior
    posterior /= posterior.sum()
    return grid, posterior

data = np.repeat([0, 1], (100, 30))
points = 1000
h = data.sum()
t = len(data) - h
grid_indicator, posterior_indicator = posterior_grid(points, h, t)
plt.figure(figsize=(8, 6))
plt.plot(grid_indicator, posterior_indicator, 'o-')
plt.title(f'heads = {h}, tails = {t}')
plt.yticks([])
plt.xlabel('Î¸')
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def posterior_grid(grid_points=50, heads=60, tails=90):
    grid = np.linspace(0, 1, grid_points)
    prior = abs(grid - 0.5)

    likelihood = stats.binom.pmf(heads, heads + tails, grid)
    posterior = likelihood * prior
    posterior /= posterior.sum()
    return grid, posterior

data = np.repeat([0, 1], (20, 6))
points = 1000
h = data.sum()
t = len(data) - h
grid_abs_diff, posterior_abs_diff = posterior_grid(points, h, t)
plt.figure(figsize=(8, 6))
plt.plot(grid_abs_diff, posterior_abs_diff, 'o-')
plt.title(f'heads = {h}, tails = {t}')
plt.yticks([])
plt.xlabel('Î¸')
plt.show()
```
